```python
with db as cur:
    cur.execute('''
    select lv.sid, tid, var_id_1, var_id_2
    from listing_variations as lv
    left join variation_phrases as vp
    on
        lv.sid = vp.sid and
        (var_id_1 = var_id or var_id_2 = var_id)
    where var_id is NULL
    and lv.sid = 'chihyu000'
    ''')
    rows = list(cur)
print len(rows)
print rows[0]
```

```python
for sid, tid, var_id_1, var_id_2 in rows:
    with db as cur:
        for locale in ('zh_TW', 'zh_CN', 'ja', 'th', 'en'):
            if var_id_1:
                print sid, var_id_1
                cur.execute(
                '''
                    insert IGNORE into variation_phrases (sid, var_id, locale, phrase, created)
                    value (%s, %s, %s, %s, NOW())
                ''', (sid, var_id_1, locale, u'[{}] {}'.format(locale, var_id_1))
                )
            if var_id_2:
                print sid, var_id_2
                cur.execute(
                '''
                    insert IGNORE into variation_phrases (sid, var_id, locale, phrase, created)
                    value (%s, %s, %s, %s, NOW())
                ''', (sid, var_id_2, locale, u'[{}] {}'.format(locale, var_id_2))
                )
            if not var_id_2:
                print sid, var_id_2
                cur.execute(
                '''
                    insert IGNORE into variation_phrases (sid, var_id, locale, phrase, created)
                    value (%s, %s, %s, %s, NOW())
                ''', (sid, var_id_2, locale, u'')
                )

    
```
